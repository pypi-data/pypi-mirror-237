## 1 实例化OkxSPOT

```python
okxSPOT = OkxSPOT(
    key="xxx",
    secret="xxx",
    passphrase="xxx",
    # 使用http和https代理，proxies={'http':'xxxxx','https:':'xxxxx'}，通requests中的proxies参数规则相同
    proxies={},
    # 转发：需搭建转发服务器，可参考：https://github.com/pyted/okx_resender
    proxy_host=None, 
)
```

## 2 获取未成交订单列表

```python
get_orders_pending = okxSPOT.trade.get_orders_pending()

get_orders_pending
```

## 3 获取未成交的开仓订单列表（购买）

```python
get_orders_pending_open = okxSPOT.trade.get_orders_pending_open()

get_orders_pending_open
```
