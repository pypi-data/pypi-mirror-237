```python
import pandas as pd
import numpy as np
from buckaroo.buckaroo_widget import BuckarooWidget
```

```python
df = pd.read_html("/Users/paddy/Downloads/bikesharing-systems-wikipedia.html")[0]
w = BuckarooWidget(df)
w
```

```python
active_bikeshares = df[df['Discontinued'].fillna("false") == "false"]
```

```python
BuckarooWidget(active_bikeshares)
```

```python
def safe_int(x):
    try:
        return float(x)
    except:
        return np.nan
def clean(df):
    df['Bicycles'] = df['Bicycles'].apply(safe_int)
    df['Stations'] = df['Stations'].apply(safe_int)
    grps = df.groupby('System')
    df_contents = {}
    df_contents['Stations'] = grps['Stations'].apply(lambda x: x.sum())
    df_contents['Bicycles'] = grps['Bicycles'].apply(lambda x: x.sum())
    df_contents['City'] = grps['City'].apply(lambda x: x.count())
    df = pd.DataFrame(df_contents)
    return df
```

```python
cleaned_df = clean(active_bikeshares)
```

```python
BuckarooWidget(cleaned_df.sort_values('Bicycles', ascending=False))
```

```python
cleaned_df.sort_values('City', ascending=False)
```
